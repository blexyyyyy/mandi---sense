```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['Date'])
df = df.sort_values('Date')

df['Date'].min(), df['Date'].max()
```

```python
print("Top commodities:")
print(df['Commodity'].value_counts().head(10))

print("\nTop states:")
print(df['State'].value_counts().head(10))

print("\nExample markets:")
print(df['Market'].value_counts().head(10))
```

```python
# Fix extra spaces
df['State'] = df['State'].str.strip()
df['Commodity'] = df['Commodity'].str.strip()
df['Market'] = df['Market'].str.strip()
df['Variety'] = df['Variety'].str.strip()

# Filter for Onion + Maharashtra
filtered = df[(df['Commodity'] == "Onion") & (df['State'] == "Maharashtra")]

print("Filtered shape:", filtered.shape)
filtered.head()
```

```python
filtered.to_csv("../data/processed/onion_maharashtra_filtered.csv", index=False)
print("Saved filtered dataset.")
```
